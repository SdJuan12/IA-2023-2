```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("C:\\Users\\ADMIN\\OneDrive - Universidad Nacional de Colombia\\iris.csv")

class AnalizadorDatosIris:
    def __init__(self, dataframe):
        self.data = dataframe

    def estadisticas_basicas(self):
        return self.data.describe()

    def nombres_variables(self):
        return self.data.columns.tolist()

    def analizar_variable(self, nombre_variable):
        if nombre_variable in self.data.columns:
            variable = self.data[nombre_variable]
            media = variable.mean()
            mediana = variable.median()
            desviacion = variable.std()
            q25 = variable.quantile(0.25)
            q50 = variable.quantile(0.50)
            q75 = variable.quantile(0.75)

            plt.hist(variable, bins=20)
            plt.xlabel(nombre_variable)
            plt.ylabel("Frecuencia")
            plt.title(f"Histograma de {nombre_variable}")
            plt.show()

            return {
                "Media": media,
                "Mediana": mediana,
                "Desviación Estándar": desviacion,
                "Percentil 25": q25,
                "Percentil 50": q50,
                "Percentil 75": q75
            }
        else:
            return "La variable especificada no existe en el conjunto de datos."

if __name__ == "__main__":
    analizador = AnalizadorDatosIris(data)
    print("Estadísticas básicas:")
    print(analizador.estadisticas_basicas())
    print("\nNombres de las variables disponibles:")
    print(analizador.nombres_variables())
    print("\nAnálisis de la variable 'sepal.length':")
    print(analizador.analizar_variable("sepal.length"))
```
